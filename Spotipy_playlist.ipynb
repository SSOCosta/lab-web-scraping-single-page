# EXTRACTING FIRST PLAYLIST

```python
#Making a connection to spotipy which connects us to spotify
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['clientid'],
                                                           client_secret=secrets_dict['clientsecret']))
```

```python
# Looking at the playlist content
playlist = sp.user_playlist_tracks("spotify", "6yPiKpy7evrwvZodByKvM9")
#playlist
```

```python
# total number of tracks 
playlist["total"] 
```

```python
# Checking the dictionary key values
playlist.keys() 
```

```python
# Each page length
len(playlist["items"])
```

```python
# Looking at the dictionary inside 'items' were what we need looks to be
playlist["items"][0].keys()
```

```python
# Looking at subset 'track' in 'items'
playlist["items"][0]["track"]
```

```python
# Using a function to have the playlist of tracks
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000) 
    return tracks
```

```python
all_tracks = get_playlist_tracks("6yPiKpy7evrwvZodByKvM9")
len(all_tracks)
```

```python
import pandas as pd
from pandas import json_normalize
tracks = json_normalize(all_tracks)
```

```python
# Checking my dataframe
tracks.head()
```

```python
# Checking for tracks key alues
tracks.keys()
```

```python
def expand_list_dict(row):
    df = json_normalize(row['track.artists'])
    df['song_id'] = row['track.id']
    return df

tracks['artists_dfs'] = tracks.apply(expand_list_dict, axis=1)
tracks['artists_dfs'][0]
```

```python
# now we create a new dataframe with all these artists
artist_df = pd.DataFrame(columns=['href', 'id', 'name', 'type', 'uri'])
for mini_df in tracks['artists_dfs']:
    #display(val)
    artist_df = pd.concat([artist_df, mini_df], axis=0)
    
artist_df
```

```python
# now we merge (join) the two dataframes, keeping only the fields we need
df_merged = pd.merge(left=tracks,
                    right=artist_df,
                    how='inner',
                    left_on='track.id',
                    right_on='song_id')
df_merged.head()
```

```python
df_final = df_merged[['track.name', 'name', 'song_id']]
df_final
```

```python
# pulling audio features for each page of 100 at a time
chunks = [(i, i+100) for i in range(0, len(df_final), 100)]
chunks
audio_features_list = []
df_final['song_id'] = df_final['song_id'].astype(str)
for chunk in chunks:
    id_list100 = df_final['song_id'][chunk[0]:chunk[1]]
    audio_features_list = audio_features_list + sp.audio_features(id_list100)
    sleep(randint(1,3000)/1000)
len(audio_features_list)
```

```python
audio_features_df = json_normalize(audio_features_list)
```

```python
df_w_audio_ft = pd.merge(left=df_final,
                        right=audio_features_df,
                        how='inner',
                        left_on='song_id',
                        right_on='id')
df_w_audio_ft
```
